# Report: We Rate Dogs 

The WeRateDogs tweets dataset contained names, ratings of dogs etc. as well as favorite counts and retweet counts of each of the tweets.

The following Insights were drawn from the analysis of the data

## Insights

#### 1. Most Popular Dogs Name

The most popular Dog name is Charlie with about 11 counts. It is Closely followed by Cooper, Oliver and Lucy with the number of occurrence been 10. While the least popular dog name was Lola.

#### 2. Is there any correlation or Relationship  between Favorite and Retweet?

There is a positive relationship between favorite tweet and the retweet with a correlation co-efficient of 0.712. This goes to show that there is a high likehood of favorite tweets been retweeted.

###### 3. What is the Most Used Twitter Source?

From the Visualization, it shows that most users make use of twitter for iPhone.
